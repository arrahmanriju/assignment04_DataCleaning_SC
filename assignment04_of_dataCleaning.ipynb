**5. Outlier Detection & Handling**

---

Problem Detection


Methods:

- Boxplot
- IQR
- Z-Score
- Domain Rules

Example:

Age < 0
Salary > 10M

### Solutions

- Remove
- Cap (Winsorization)
- Log Transform


````python
#for experience year column

Q1 = df["Experience_Years"].quantile(0.25)
Q3 = df["Experience_Years"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["Experience_Years"] < lower) |
    (df["Experience_Years"] > upper)
]

outliers
print(len(outliers))
````

````python
#for Working_Hours_Per_Week columns
Q1 = df["Working_Hours_Per_Week"].quantile(0.25)
Q3 = df["Working_Hours_Per_Week"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["Working_Hours_Per_Week"] < lower) |
    (df["Working_Hours_Per_Week"] > upper)
]

outliers
print(len(outliers))
````

````python
#for Injury_History columns
Q1 = df["Injury_History"].quantile(0.25)
Q3 = df["Injury_History"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["Injury_History"] < lower) |
    (df["Injury_History"] > upper)
]

outliers
print(len(outliers))
````

````python
#Remove outliers from Injury History columns

df_task_five = df.copy()
df_task_five = df_task_five[
    (df['Injury_History'] >= lower) &
    (df['Injury_History'] <= upper)
]
````

````python
#cap for Injury History columns
df_task_five['Injury_History'] = df_task_five['Injury_History'].clip(lower, upper)
````

````python
#Log Transform for Injury History Columns

df_task_five['Injury_History'] = np.log(df_task_five['Injury_History'])
````

````python
print(dfLastCopy["Injury_History"].value_counts())
print(dfLastCopy["Injury_History"].value_counts(normalize=True) * 100)
````

````python

# ---------------------
# Oversampling
# ---------------------

from sklearn.utils import resample

majority = dfLastCopy[dfLastCopy.Injury_History == 0]
minority = dfLastCopy[dfLastCopy.Injury_History == 1]

minority_up = resample(
    minority,
    replace=True,
    n_samples=len(majority),
    random_state=42
)

df_balanced = pd.concat(
    [majority, minority_up]
)
````

````python
# ---------------------
# Undersampling
# ---------------------

majority_down = resample(
    majority,
    replace=False,
    n_samples=len(minority),
    random_state=42
)

df_under = pd.concat(
    [majority_down, minority]
)
````
